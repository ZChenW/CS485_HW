# Demonstrating part-of-speech tagging

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentence = "The window was open."
words = nltk.word_tokenize(sentence)
print(words)
tagged_words = nltk.pos_tag(words)
for w in tagged_words: print(w)
```

```python
sentence = "Can you open the window?"
words = nltk.word_tokenize(sentence)
print(words)
tagged_words = nltk.pos_tag(words)
for w in tagged_words: print(w)
```

```python
sentence = "They refuse to permit us to obtain the refuse permit"
# this odd sentence is from the NLTK book, https://www.nltk.org/book/ch05.html
words = nltk.word_tokenize(sentence)
print(words)
tagged_words = nltk.pos_tag(words)
for w in tagged_words: print(w)
```
